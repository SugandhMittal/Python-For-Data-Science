```python
#Question 1
#For a) Signed Char

import array
import random

# Define the typecode for signed char
typecode = 'b'

# Define the length of the array
array_length = 10

# Create an empty array of signed char
random_array = array.array(typecode)

# Populate the array with random signed char values
for _ in range(array_length):
    random_value = random.randint(-128, 127)  # Signed char range is -128 to 127
    random_array.append(random_value)

# Print the random array
print(random_array)
```

```python
# Array operations
# Traverse (Print Elements):
def traverse_array(arr):
    for element in arr:
        print(element, end=" ")
    print()  # Print a newline for formatting

print("Traversing the random array:")
traverse_array(random_array)

# Insertion:
def insert_element(arr, index, value):
    arr.insert(index, value)

insert_element(random_array, 3, -5)
print("After inserting -5 at index 3:")
traverse_array(random_array)

# Deletion:
def delete_element(arr, index):
    if 0 <= index < len(arr):
        arr.pop(index)
    else:
        print("Index out of range")

delete_element(random_array, 6)
print("After deleting element at index 6:")
traverse_array(random_array)

# Search:
def search_element(arr, target):
    for index, element in enumerate(arr):
        if element == target:
            return index
    return -1

search_value = -5
result = search_element(random_array, search_value)
if result != -1:
    print(f"{search_value} found at index {result}")
else:
    print(f"{search_value} not found")

# Update:
def update_element(arr, index, new_value):
    if 0 <= index < len(arr):
        arr[index] = new_value
    else:
        print("Index out of range")

update_element(random_array, 2, 99)
print("After updating element at index 2 to 99:")
traverse_array(random_array)
```

```python
#Question 5

import numpy as np

# (1) Create a NumPy array of 10 random integers from 0 to 100
random_integers = np.random.randint(0, 101, 10)

# (2) Calculate the mean
mean = np.mean(random_integers)

# (3) Create an array of type float based on it, and replace outliers
# (values < mean-std or > mean+std) with the mean
std_deviation = np.std(random_integers)
float_array = random_integers.astype(float)
float_array[(float_array < (mean - std_deviation)) | (float_array > (mean + std_deviation))] = mean

# (4) Create a NumPy array of 20 random floats from 0 to 100
random_floats = np.random.uniform(0, 100, 20)

# (5) Replace those numbers whose integer portion is even with the mean
integer_portions = np.floor(random_floats)
even_mask = (integer_portions % 2 == 0)
random_floats[even_mask] = mean

# Print the results
print("(1) Random Integers:", random_integers)
print("(2) Mean of Random Integers:", mean)
print("(3) Float Array with Outliers Replaced:", float_array)
print("(4) Random Floats:", random_floats)


```

```python
#Question 7

import numpy as np

# Create an empty list to store the 1D arrays
arrays_list = []

# Create 10 random 1D arrays of size 10 and append them to the list
for _ in range(10):
    random_array = np.random.rand(10)  # Create a random 1D array
    arrays_list.append(random_array)

# Stack the 1D arrays together to create a 2D array
stacked_array = np.vstack(arrays_list)

# Normalize the data (scale between 0 and 1)
min_value = np.min(stacked_array)
max_value = np.max(stacked_array)
normalized_array = (stacked_array - min_value) / (max_value - min_value)

# Reshape the 2D array back into a 1D array
reshaped_array = normalized_array.flatten()

# Print the results
print("(1) List of 1D Arrays:")
for i, arr in enumerate(arrays_list):
    print(f"Array {i + 1}:", arr)

print("\n(2) Stacked 2D Array:")
print(stacked_array)

print("\n(3) Normalized 2D Array:")
print(normalized_array)

print("\n(4) Reshaped 1D Array:")
print(reshaped_array)
```

```python
#Question 12

import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(0, 2 * np.pi, 100)  # Values from 0 to 2*pi

# Calculate the sine and cosine values
sin_wave = np.sin(x)
cos_wave = np.cos(x)

# Plot both waves in the same graph
plt.figure(figsize=(8, 4))
plt.plot(x, sin_wave, label='sin(x)', color='b')
plt.plot(x, cos_wave, label='cos(x)', color='r')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Set axis limits
plt.xlim(0, 2 * np.pi)
plt.ylim(-1.2, 1.2)

# Annotate some points on the graph
plt.annotate('Maximum', xy=(np.pi / 2, 1), xytext=(np.pi / 2, 1.2),
             arrowprops=dict(arrowstyle='->'), fontsize=10, color='b')
plt.annotate('Minimum', xy=(3 * np.pi / 2, -1), xytext=(3 * np.pi / 2, -1.2),
             arrowprops=dict(arrowstyle='->'), fontsize=10, color='r')

# Show the combined graph
plt.show()

# Plot the same graphs separately using subplots
plt.figure(figsize=(10, 5))

# Subplot 1: Sine wave
plt.subplot(2, 1, 1)
plt.plot(x, sin_wave, color='b')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

# Subplot 2: Cosine wave
plt.subplot(2, 1, 2)
plt.plot(x, cos_wave, color='r')
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
#Question 14

import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y)
def f(x, y):
    return (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

# Generate a grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Calculate the function values for the grid
Z = f(X, Y)

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=15, cmap='viridis')  # Adjust the number of levels as needed
plt.colorbar(contour, label='f(x, y)')
plt.title('Contour Plot of f(x, y)')
plt.xlabel('x')
plt.ylabel('y')

# Show the contour plot
plt.show()
```

```python
#Question 15

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Calculate the Gaussian function values for the grid
Z = np.exp(-(X**2 + Y**2) / 2) / (2 * np.pi)

# Create a 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D Gaussian surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add color bar
fig.colorbar(surf, label='Gaussian Value')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Gaussian Value')
ax.set_title('3D Gaussian Plot')

# Show the 3D plot
plt.show()
```

```python
#Plotting of interpolated line graph for the above data

import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

# Original data
x = np.linspace(0, 5, 10)
y = np.cos(x**2/3 + 4)

# Create interpolation functions for linear and cubic interpolation
linear_interp = interpolate.interp1d(x, y, kind='linear')
cubic_interp = interpolate.interp1d(x, y, kind='cubic')

# Define a finer grid of x values for plotting the interpolated curves
x_fine = np.linspace(0, 5, 100)

# Calculate interpolated values using both methods
y_linear_interp = linear_interp(x_fine)
y_cubic_interp = cubic_interp(x_fine)

# Plot the original data and interpolated curves
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='Original Data', markersize=8)
plt.plot(x_fine, y_linear_interp, 'g-', label='Linear Interpolation')
plt.plot(x_fine, y_cubic_interp, 'b--', label='Cubic Interpolation')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolation with Linear and Cubic Methods')
plt.grid(True)
plt.show()
```
