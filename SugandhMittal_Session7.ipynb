# **Web Scraping on Flipkart**

```python
import requests
from bs4 import BeautifulSoup

# Function to scrape and print data from a Flipkart search page
def scrape_flipkart_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract and print the webpage title
    webpage_title = soup.title.string
    print("Webpage Title:", webpage_title)

    # Extract and print headers
    headers = [header.text for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
    print("Headers:", headers)

    # Extract and print HTML tags
    tags = {tag.name for tag in soup.find_all()}
    print("HTML Tags:", tags)

    # Extract and print text consisting of at least 2 complete sentences
    paragraphs = soup.find_all('p')
    sentences = [p.text.strip() for p in paragraphs if '.' in p.text and len(p.text.split('.')) > 1]
    print("Text with 2+ Sentences:", sentences)

    # Find and print links within the page
    links = [a['href'] for a in soup.find_all('a', href=True)]
    print("Links on the Page:", links)

# Main URL for Flipkart mobile search (you can change this)
url = "https://www.flipkart.com/search?q=mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobile%7CMobiles&requestId=d1835531-1479-425f-a59d-456fac40318f&as-searchtext=mobile"

# Call the function to scrape and print data from the specified URL
scrape_flipkart_page(url)

```

# **For W3 schools**

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.w3schools.com/python/default.asp"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Question1: Get the webpage title and headers
webpage_title = soup.title.string
headers = [header.text for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
print("Webpage Title:", webpage_title)
print("Headers:", headers)

# Question2: Get all the tags in the HTML
tags = {tag.name for tag in soup.find_all()}
print("HTML Tags:", tags)
print()

# Question3: Get text consisting of at least 2 complete sentences
paragraphs = soup.find_all('p')
sentences = [p.text.strip() for p in paragraphs if '.' in p.text and len(p.text.split('.')) > 1]
print("Text with 2+ Sentences:", sentences)

# Step 4: Find links within the page and scrape the titles of the new webpages
link = soup.find_all('a')[0]["href"]
response_new = requests.get(link)
soup_new = BeautifulSoup(response_new.text, 'html.parser')
webpage_title_new = soup_new.title.string
print("New Webpage Title:", webpage_title_new)
```

```python
import requests
from bs4 import BeautifulSoup

# Initialize a list to store the scraped content
scraped_content = []

# Scrape 3 pages
for page_number in range(1, 4):
    url = f"https://www.w3schools.com/python/default.asp?page={page_number}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the content you want from the page, e.g., paragraphs with class "w3-container"
    content = soup.find_all('div', class_='w3-container')

    # Append the scraped content to the list
    scraped_content.extend(content)

# Print the scraped content
for idx, content in enumerate(scraped_content, start=1):
    print(f"Page {idx} Content:")
    print(content.get_text())
    print("=" * 50)
```
